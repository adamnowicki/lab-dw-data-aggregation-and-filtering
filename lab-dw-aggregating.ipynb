1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")

df.head()

def clean_my_columns(df):
    df.columns = df.columns.str.replace(" ", "_").str.lower().str.strip()
    return df

clean_my_columns(df)
```

```python
df=df.drop(['unnamed:_0'], axis=1)
```

```python
df1 = df.loc[(df['customer_lifetime_value'] > 1000) & (df['response'] == "Yes")]
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python

display(df[df['response'] == "Yes"].groupby(['policy_type', 'gender'])['total_claim_amount'].mean())

# Women have higher total claim amounts than men.
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
display(df.groupby(['education', 'gender'])['customer_lifetime_value'].min().round(0))

display(df.groupby(['education', 'gender'])['customer_lifetime_value'].max().round(0))

display(df.groupby(['education', 'gender'])['customer_lifetime_value'].median().round(0))


# Lower educated clients have the highest lifetime value (probably due to the earlier age of buying policies)
# Females with doctorate have the lowest CLV 
# Male with lower education have the highest CLV (median) and the highest recored value (83325)
```
